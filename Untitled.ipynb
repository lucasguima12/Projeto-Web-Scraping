```python
import re
import pandas as pd
from bs4 import BeautifulSoup
import requests

def criar_tabela(ano):
    url_musica = f'https://maistocadas.mus.br/{ano}/'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }
    
    try:
        response = requests.get(url_musica, headers=headers)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")
        return pd.DataFrame()

    soup = BeautifulSoup(response.text, 'html.parser')
    target_div = soup.find('div', {'class': 'lista'})
    if target_div is None:
        print('Target div not found')
        return pd.DataFrame()

    target_div_text = target_div.text.replace(f"Lista das mais tocadas {ano} ", "")
    
    # Adjusted regex pattern to handle the case where there's no space before the dash
    pattern = re.compile(r'(.*?)(?: –| -)(.*?)(?: –| -) Tocou (.*?) vezes(?: \((.*?)\))?')
    matches = pattern.finditer(target_div_text)
    
    data = []
    for match in matches:
        song, artist, times_played, recorder = match.groups()
        times_played = int(times_played.replace('.', ''))
        data.append({
            'Música': song.strip(),
            'Artista': artist.strip(),
            'Gravadora': recorder.strip() if recorder else 'Unknown',
            'Vezes Tocada': times_played
        })

    if not data:
        print('No songs found')
        return pd.DataFrame()

    df = pd.DataFrame(data)
    return df
```
